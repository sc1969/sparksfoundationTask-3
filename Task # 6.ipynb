## >>In the beginning, the whole data is considered as the root, thereafter, we use the algorithms to make a split or divide the root into subtrees.

## >>The feature values are considered to be categorical. If the values are continuous, then they are separated prior to building the model.

## >>Records are distributed recursively on the basis of attribute values.

# Importing the modules

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

# Getting the Data 

```python
iris = load_iris()
x = pd.DataFrame(iris.data, columns=iris.feature_names)
```

# Viewing the Predictor Data

```python
x.head(5)
```

# Viewing the Target Data

```python
y = iris.target
y
```

# Splitting the data

```python
# Splitting the Data:
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size =0.25, random_state = 0)
```

# Creating the model 

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

# Checking the score (accuracy) of the model & Confusion Matrix

```python
from sklearn.metrics import accuracy_score
y_new =model.predict(X_test)
print(accuracy_score(y_test, y_new))
```

```python
from sklearn.metrics import plot_confusion_matrix 
plot_confusion_matrix(model, X_test, y_test , display_labels = ["Setosa", "VertiColor", "Virginica"], cmap = plt.cm.Blues)
plt.title("Plotting the Confusion Matrix")
plt.show()
```

# Visualizing the Decision Tree Using Matplotlib

```python
feature_name=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
class_name=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (20,20))
out = tree.plot_tree(model, feature_names = feature_name, class_names = class_name, filled= True)
for o in out:
    arrow = o.arrow_patch
    if arrow is not None:
        arrow.set_edgecolor('black')
        arrow.set_linewidth(3)
```
